```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 25]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

plt.title("Simple line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, [5, 10, 15, 10, 5])
plt.show()
```

```python
plt.figure()
plt.plot(x, y)
plt.show()

plt.figure()
plt.plot(x, [5, 10, 15, 10, 5])
plt.show()
```

```python
plt.plot(x, y, label="First Line")
plt.plot(x, [5, 10, 15, 10, 5], label="Second Line")
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 25]
plt.plot(x, y, color="red")
plt.show()
```

```python
plt.plot(x, y, linestyle="--")
plt.show()
```

```python
plt.plot(x, y, linewidth=3)
plt.show()
```

```python
plt.plot(x, y, color="blue", linestyle=":", linewidth=2)
plt.title("Styled Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, color="blue")
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
names = ["Ahmed", "Sara","Omar"]
scores = [85, 90, 78]
```

```python
plt.bar(names, scores)
plt.show()
```

```python
plt.bar(names, scores, color="green")
plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()
```

```python
scores = [50, 60, 70, 70, 80, 85, 90, 90, 90, 100]
plt.hist(scores)
plt.show()
```

```python
plt.hist(scores, bins=5)
plt.show()
```

```python
plt.hist(scores, bins=5)
plt.title("Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

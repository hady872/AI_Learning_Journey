```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 20, 25]
y2 = [5, 10, 15, 10, 5]

# Figure size
plt.figure(figsize=(8, 5))

# First line
plt.plot(x, y1,
         color="blue",
         marker="o",
         label="Line 1")

# Second line
plt.plot(x, y2,
         color="red",
         linestyle="--",
         marker="s",
         label="Line 2")

# Labels and title
plt.title("Professional Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend and grid
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 25]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

plt.title("Simple line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, [5, 10, 15, 10, 5])
plt.show()
```

```python
plt.figure()
plt.plot(x, y)
plt.show()

plt.figure()
plt.plot(x, [5, 10, 15, 10, 5])
plt.show()
```

```python
plt.plot(x, y, label="First Line")
plt.plot(x, [5, 10, 15, 10, 5], label="Second Line")
plt.legend()
plt.show()
```

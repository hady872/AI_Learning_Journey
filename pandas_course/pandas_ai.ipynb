#Day 8: Filtering & Conditions in Pandas

```python
import pandas as pd
data = {
    "name": ["Ahmed", "Sara", "Omar", "Lina"],
    "math": [90, 85, 85, 70],
    "english": [88, 92, 85, 75],
}
df = pd.DataFrame(data)
df
```

```python
df[df["math"] > 80]
```

```python
df[df["name"] =="Ahmed"]
```

```python
df[(df["math"] > 80) & (df["english"] > 85)]
```

```python
df[(df["math"] > 70) | (df["english"] > 80)]
```

```python
df[df["math"].isin(["Ahmed", "Lina"])]
```

```python
df[df["math"] > 70].sort_values(by="math",ascending=False)
```

```python
df[df["english"] >=80]
```

```python

df[df["english"] >=80].sort_values(by="english",ascending=False)
```

```python
df.sort_values(by="english",ascending=False)
```

#Day 9: GroupBy

```python
import pandas as pd
data = {
    "student": ["Ahmed", "Sara", "Omar", "Lina", "Mona"],
    "department": ["CS", "IT", "CS", "IT", "CS"],
    "score": [90, 85, 88, 92, 95]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby("department")["score"].mean()
```

```python
df.groupby("department")["score"].agg(['mean', 'max', 'min','count'])
```

```python
df.groupby(["department", "student"])["score"].mean()
```

```python
result = df.groupby("department")["score"].mean().reset_index()
result
```

Day 11: GroupBy & Aggregation

```python
import pandas as pd

data = {
    "department": ["IT", "HR", "IT", "HR", "Sales","Sales"],
    "employee": ["Ahmed", "Sara", "Omar", "Lina", "Mona", "Ali"],
    "salary": [5000, 4500, 5200, 4800, 5500, 5300]
}
df = pd.DataFrame(data)
df
```

```python
df.groupby("department")["salary"].mean().sort_values(ascending=False)
```

```python
result = df.groupby("department")["salary"].mean().reset_index()
result
```

```python
df.groupby("department").agg({"salary":["mean", "sum"], "employee":"count"})
```

```python
df.groupby(["department","employee"]).agg(["mean", "sum", "max", "min"])
```

```python
df.groupby("department")["salary"].mean()
```

```python
df.groupby("department")["salary"].agg(["mean", "sum", "max", "min"])
```

Day 12: Merge & Join

```python
import pandas as pd

customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Ahmed", "Sara", "Omar", "Lina"]
})
orders = pd.DataFrame({
    "customer_id": [1, 1, 2, 3],
    "order_amount": [500, 300, 700, 200]
})
customers
```

```python
orders
```

```python
merged = pd.merge(customers, orders, on="customer_id")
merged
```

```python
left_join = pd.merge(customers, orders, on="customer_id", how="left")
left_join
```

```python
right_join = pd.merge(customers, orders, on="customer_id", how="right")
right_join
```

```python
outer_join = pd.merge(customers, orders, on="customer_id", how="outer")
outer_join
```

```python
merged.groupby("name")["order_amount"].sum()
```

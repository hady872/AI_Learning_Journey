```python
import pandas as pd
import numpy as np

data = {
    "name": ["Ahmed", "Sara", "Omar", "Lina", "Mona"],
    "math": [np.nan, 85, np.nan, 70, 95],
    "english": [88, 93, 75, np.nan, 90],
    "science": [84, 80, 78, 88, np.nan],
    "biology": [82, 84, 89, 87, 83]
}

df = pd.DataFrame(data)
df
```

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
df["total"] = df[["math", "english", "science"]].sum(axis=1)
df["average"] = df[["math", "english", "science"]].mean(axis=1)
df
```

```python
df.sort_values(by="average", ascending=False)
```

```python
df.mean(numeric_only=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(df["name"], df["average"], color="blue")

plt.title("Average Score per Student")
plt.xlabel("Studnets")
plt.ylabel("Average Score")

plt.show()
```

```python
plt.hist(df["average"], bins=5)

plt.title("Distrubution of Average Scores")
plt.xlabel("Average")
plt.ylabel("Frequency")

plt.show()
```

```python
import yfinance as yf
import pandas as pd

def analyze_stock(symbol):
    df = yf.download(symbol, period="1mo")

    # تحويل الأعمدة لو فيها MultiIndex
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # نتأكد إن Close عمود واحد
    close = df["Close"]

    # لو طلع DataFrame ناخد أول عمود
    if isinstance(close, pd.DataFrame):
        close = close.iloc[:, 0]

    df["Close"] = close

    # الحسابات
    df["MA_5"] = df["Close"].rolling(5).mean()
    df["Change_%"] = df["Close"].pct_change() * 100

    # آخر قيم
    last_close = float(df["Close"].dropna().iloc[-1])
    last_ma5 = float(df["MA_5"].dropna().iloc[-1])
    last_change = float(df["Change_%"].dropna().iloc[-1])
    last_volume = int(df["Volume"].dropna().iloc[-1])

    # الاتجاه
    if last_close > last_ma5:
        trend = "Above MA5 (Bullish)"
    else:
        trend = "Below MA5 (Bearish)"

    return {
        "symbol": symbol,
        "close": round(last_close, 2),
        "ma5": round(last_ma5, 2),
        "change_percent": round(last_change, 2),
        "volume": last_volume,
        "trend": trend
    }

def format_stock_message(result):
    message = f"""
Stock: {result['symbol']}
Close: {result['close']}
MA 5: {result['ma5']}
Change %: {result['change_percent']}%
Volume: {result['volume']}
Trend: {result['trend']}
"""
    return message
```

```python
result = analyze_stock("AAPL")
result
```

```python
msg = format_stock_message(result)
print(msg)
```
